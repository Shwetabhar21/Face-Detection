```python
!pip install deepface
import cv2
from google.colab.patches import cv2_imshow
from deepface import DeepFace

# Load the face detection classifier
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

# Load the image
img = cv2.imread('/content/images.jpg')

# Check if the image was loaded successfully
if img is not None:
    try:
        # Analyze the full image once
        results = DeepFace.analyze(img, actions=['age', 'gender', 'emotion'], enforce_detection=False)

        # If only one face is found, wrap it in a list
        if not isinstance(results, list):
            results = [results]

        # Perform face detection (for drawing boxes)
        faces = face_cascade.detectMultiScale(img, 1.1, 4)

        for i, (x, y, w, h) in enumerate(faces):
            # Draw face rectangle
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

            # Use the corresponding result from DeepFace
            if i < len(results):
                age = int(results[i]['age'])

                # Fix gender format
                gender_info = results[i]['gender']
                if isinstance(gender_info, dict):
                    gender = 'Female' if gender_info.get('Woman', 0) > gender_info.get('Man', 0) else 'Male'
                else:
                    gender = gender_info  # fallback if it's already a string

                emotion = results[i]['dominant_emotion']

                # Multiline label
                label_lines = [
                    f"Age: {age}",
                    f"Gender: {gender}",
                    f"Emotion: {emotion}"
                ]

                font = cv2.FONT_HERSHEY_SIMPLEX
                font_scale = 0.4
                font_thickness = 1
                line_height = 12

                # Place text safely above or below
                text_y = y - 10 if y > 40 else y + h + 10

                for j, line in enumerate(label_lines):
                    y_pos = text_y + j * line_height
                    cv2.putText(img, line, (x, y_pos), font, font_scale, (0, 255, 0), font_thickness)
    except Exception as e:
        print(f"DeepFace analysis error: {e}")

    # Show image
    cv2_imshow(img)

    # Save image
    cv2.imwrite("/content/images.jpg", img)
    print('Successfully saved as output.jpg')
else:
    print("Error: Could not load the image.")
```
